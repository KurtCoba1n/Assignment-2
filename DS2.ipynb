```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from pandas_profiling import ProfileReport
```

```python
#read the datasets for amazon and flipkart
amz=pd.read_csv('amz_com-ecommerce_sample.csv',encoding='unicode escape')
```

```python
flip=pd.read_csv('flipkart_com-ecommerce_sample.csv',encoding='unicode escape')
```

```python
#drop uniqid column
flip.drop('uniq_id',axis=1,inplace=True)
```

```python
amz.drop('uniq_id',axis=1,inplace=True)
```

```python
flip
```

```python
#create a preprocessing function to drop all the unnecessary columns 
def preprocessing(flip):
    flip.drop(['crawl_timestamp','product_url','product_category_tree','pid','image','is_FK_Advantage_product'],axis=1,inplace=True)
    flip.drop(['description','product_rating','overall_rating','brand','product_specifications'],inplace=True,axis=1)
```

```python
#apply preprocessing function on flipkart dataset
preprocessing(flip)
```

```python
flip
```

```python
#apply preprocessing function on amazon dataset
preprocessing(amz)
```

```python
amz
```

```python
#merge both the flipkart and amazon data set into a single dataframe
df=pd.concat([flip,amz],axis=1)
```

```python
df
```

```python
#create an array with the name of columns
df_cols=['product name in Flipkart','Retail price Flipkart','Discounted price flipkart','product name in Amazon','Retail price Amazon','Discounted price Amazon']
```

```python
#update the name of columns
df.columns=df_cols
```

```python
df
```

```python
#remove entries where product name in flipkar is not equal to the product name in amazon
df= df.loc[(df['product name in Flipkart']==df['product name in Amazon'])]
```
